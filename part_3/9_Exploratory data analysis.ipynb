```python
# nltk.download()

# Importing data
import nltk
from nltk.corpus import webtext
nltk.download('webtext')

wt_sentences = webtext.sents('firefox.txt')
wt_words = webtext.words('firefox.txt')

# firefox.txt predefined corpora
# NLTK's small collection of web text includes content from a Firefox 
# discussion forum  put into firefox.txt
```

```python
# Import necessary libraries
from nltk.probability import FreqDist
from nltk.corpus import stopwords
import string
```

```python
# Check number of words in the data
len(wt_sentences)
```

```python
len(wt_words)
```

```python
# Compute the frequency of all words in the reviews
frequency_dist = nltk.FreqDist(wt_words)
frequency_dist
```

```python
sorted_frequency_dist =sorted(frequency_dist,key=frequency_dist.__getitem__, reverse=True)
sorted_frequency_dist
```

```python
# Consider words with length greater than 3 and print word, frequency
## type your code 

large_words = dict([(k,v) for k,v in frequency_dist.items() if len(k)>3])



# to see the keys i.e words and their frequencies
# print(sorted(large_words, key=large_words.get, reverse=True))
s = [(k, large_words[k]) for k in sorted(large_words, key=large_words.get, reverse=True)]
for k, v in s:
     print(k, v)
```

```python
# and plot the Frequency distribution
## type your code here

frequency_dist = nltk.FreqDist(large_words)
frequency_dist.plot(25,cumulative = False)


```

On brief inspection of the data, you will see a need of data pre-processing.

Do as follows :

1> Remove the "Title" feature.

2> Remove the rows where "Review Text" were missing.

3> Clean "Review Text" column.

4> Using TextBlob to calculate sentiment polarity which lies in the range of [-1,1] where 1 means positive sentiment and -1 means a negative sentiment.

5> Create new feature for the length of the review.

6> Create new feature for the word count of the review.
#------------------------------------------------------------------


```python

# step 1 : get the data 
import pandas as pd
df = pd.read_csv('C:\Program Files\Python36\suven\Adv ML\datasets\datasets/Womens Clothing E-Commerce Reviews.csv')
df.head()

#------------------------------------------------------------------

```

```python
import warnings
warnings.filterwarnings("ignore")

# step 2 : Prepare data
# drop the less useful columns
df.drop('Unnamed: 0', axis=1, inplace=True)
df.drop('Title', axis=1, inplace=True)

# remove the rows which have blank "Review Text" 
df = df[~df['Review Text'].isnull()]

# function for Data pre-processing like replacing <br/>, <a> tags, & symbol, <> 
def preprocess(ReviewText):
    ReviewText = ReviewText.str.replace("(<br/>)", "")
    ReviewText = ReviewText.str.replace('(<a).*(>).*(</a>)', '')
    ReviewText = ReviewText.str.replace('(&amp)', '')
    ReviewText = ReviewText.str.replace('(&gt)', '')#&gt   <html> -> encoding -> &lt;html&gt;
    ReviewText = ReviewText.str.replace('(&lt)', '')
    ReviewText = ReviewText.str.replace('(\xa0)', ' ')  
    return ReviewText

from textblob import TextBlob

df['Review Text'] = preprocess(df['Review Text'])

df['polarity'] = df['Review Text'].map(lambda text: TextBlob(text).sentiment.polarity)
df['review_len'] = df['Review Text'].astype(str).apply(len)
df['word_count'] = df['Review Text'].apply(lambda x: len(str(x).split()))

```

```python
#----------------------------------------------------------

# To preview whether the sentiment polarity score works, 
# we randomly select 5 reviews with the highest sentiment polarity score (1):
print('5 random reviews with the highest positive sentiment polarity: \n')
cl = df.loc[df.polarity == 1, ['Review Text']].sample(5).values
for c in cl:
    print(c[0])
```

```python
# Then randomly select 5 reviews with the most neutral 
# sentiment polarity score (zero):

print('5 random reviews with the most neutral sentiment(zero) polarity: \n')
cl = df.loc[df.polarity == 0, ['Review Text']].sample(5).values
for c in cl:
    print(c[0])
```

```python
# There were only 2 reviews with the most negative sentiment polarity score:

print('2 reviews with the most negative polarity: \n')
cl = df.loc[df.polarity == -0.97500000000000009, ['Review Text']].sample(2).values
for c in cl:
    print(c[0])
```

Quick helpful tip : ReviewText = ReviewText.str.replace('(\xa0)', ' ')

replaces the non-breaking space in Latin1 with unicode space.

Text-processing software typically assumes that an automatic line break may be inserted anywhere a space character occurs; a non-breaking space prevents this from happening (provided the software recognizes the character). For example, if the text "100 km" will not quite fit at the end of a line, the software may insert a line break between "100" and "km". An editor who finds this behaviour undesirable may choose to use a non-breaking space between "100" and "km". This guarantees that the text "100 km" will not be broken: if it does not fit at the end of a line, it is moved in its entirety to the next line. https://word.tips.net/T001081_Inserting_a_Non-Breaking_Space.html
